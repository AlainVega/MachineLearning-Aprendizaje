# Red neuronal que pasa celsius a fahrenheit

```python
import pandas as pd

datos = pd.read_csv('datosEntrenamientoC2F.csv')
```

```python
datos.head()
```

```python
# obtenemos la caracteristica y la etiqueta
x = datos['celsius'].values 
y = datos['fahrenheit'].values
```

```python
import tensorflow as tf

modelo = tf.keras.Sequential([
    tf.keras.layers.Input(shape=(1,)), # la entrada, los celsius: 1 solo dato
    tf.keras.layers.Dense(units=1) # la salida, los fahrenheit
])

modelo.compile(
    optimizer=tf.keras.optimizers.Adam(.1),
    loss='mean_squared_error'
)
```

```python
entrenamiento = modelo.fit(x, y, epochs=5000)
```

```python
import matplotlib.pyplot as plt

plt.xlabel("# Epoca")
plt.ylabel("Magnitud de la perdida")
plt.plot(entrenamiento.history['loss'])
```

```python
c = float(input('Ingrese los grafos celsius: '))
prediccion = modelo.predict([c])
print(f'Prediccion: {c} celsius son: {prediccion[0]} fahrenheit')
```
